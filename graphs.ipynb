# Script to Turn All Graphs Into One-Page Documents for Easy Visualization

```python
import os
import matplotlib.pyplot as plt

# path to the directory containing graphs
all_interventions_path = 'all_interventions/graphs'
expert_systems_path = 'expert_systems/graphs'
robots_smart_devices_path = 'robots_smart_devices/graphs'

def create_big_graph_file(directory_path, output_name):
    if os.path.isdir(directory_path):
        # get list of files in the directory
        files = os.listdir(directory_path)

        # create a figure with multiple subplots
        _, axs = plt.subplots(5, 4, figsize=(20, 20))

        # flatten axs into a 1D array
        axs = axs.flatten()

        # iterate over files and plot them
        for i, file in enumerate(files):
            # load and plot each graph
            graph_path = os.path.join(directory_path, file)
            img = plt.imread(graph_path)
            axs[i].imshow(img)
            axs[i].axis('off')

        # adjust layout
        plt.tight_layout()

        # save the output to a file named 'all-graphs.png'
        plt.savefig(output_name + '.png')

        # show the plot
        plt.show()
    else:
        print("Cannot create graphs!")
```

```python
if not os.path.isdir("graphs_combined"):
    os.mkdir("graphs_combined")
```

```python
create_big_graph_file(all_interventions_path, "./graphs_combined/all_interventions_combined_graphs")
```

```python
create_big_graph_file(expert_systems_path, "./graphs_combined/expert_systems_combined_graphs")
```

```python
create_big_graph_file(robots_smart_devices_path, "./graphs_combined/robots_smart_devices_combined_graphs")
```
